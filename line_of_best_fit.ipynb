I used the concept of finding a least-squares line as an example of an optimization problem in a Calculus I course. To keep things at just one variable, we only considered lines through the origin.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Randomly generates 90 points from a multivariate normal distribution to simulate test scores for 90 students
#from two tests, and plots the points.

np.random.seed(seed=37)
g = np.random.multivariate_normal([75, 70], [[100, 80], [80, 100]], 90)
x = g[:,0]
y = g[:,1]

%matplotlib inline
plt.scatter(x,y);
plt.xlabel('Test 1 Score');
plt.ylabel('Test 2 Score');
plt.title('Test 1 and 2 Scores for 90 Students');
plt.xlim([50,100]);
plt.ylim([40,100]);
```

```python
#The direct calculation of the slope of the line through the origin that minimizes the mean squared error
#(the formula of which was derived using Calculus I techniques).

m = np.sum(x*y)/np.sum(x**2)
m
```

```python
#Plots the points again with the least-squares line through the origin.

plt.scatter(x, y)
plt.xlabel('Test 1 Score');
plt.ylabel('Test 2 Score');
plt.title('Test 1 and 2 Scores for 90 Students');
plt.xlim([50,100]);
plt.ylim([40,100]);
plt.plot([49, 100], [m*49, m*100], label='Line through origin of best fit: y = 0.9299x', color='xkcd:burnt orange');
plt.legend();
```

```python
#Computes the unrestricted least-squares line.

m, b = np.polyfit(x, y, 1)
print(m, b)
```

```python
#Graphs the points with the unrestricted least-squares line.

plt.scatter(x, y)
plt.xlabel('Test 1 Score');
plt.ylabel('Test 2 Score');
plt.title('Test 1 and 2 Scores for 90 Students')
plt.xlim([50,100])
plt.ylim([40,100])
plt.plot([49,100], [m*49+b, m*100+b], label='Line of best fit: y = 0.7588x + 12.961', color='xkcd:burnt orange');
plt.legend();
```
